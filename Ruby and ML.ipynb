## Linear regression

```ruby
boston_df = Daru::DataFrame.from_csv 'boston_house_prices.csv'
boston_df.vectors = Daru::Index.new([:crim,:zn,:indus,:chas,:nox,:rm,:age,:dis,:rad,:tax,:ptratio,:b,:lstat,:medv])

x_data = boston_df[:rm].data.to_a.collect { |i| [i.to_i] }
y_data = boston_df[:medv].data.to_a.collect { |i| i.to_i }


linear_regression = RubyLinearRegression.new
linear_regression.load_training_data(x_data, y_data)

time = Benchmark.measure do
  (0..10).each {
    linear_regression.train_gradient_descent(0.0005, 1, false)
  }
end
puts (time.real/10)*1000

linear_regression.train_gradient_descent(0.0005, 1, false)

test_data = [6]

time = Benchmark.measure do
  (0..10).each {
    predicted = linear_regression.predict(test_data)
  }
end

puts (time.real/10)*1000
```

## Decision trees

```ruby
require 'decisiontree'
require 'datasets'

iris = Datasets::Iris.new
iris.each do |record|
  p [
        record.sepal_length,
        record.sepal_width,
        record.petal_length,
        record.petal_width,
        record.class,
    ]
end

attributes = ['sepal_length','sepal_width','petal_lenght', 'petal_width']
iris_table = iris.to_table.to_a

sepal_length=iris_table[0][1]
sepal_width =iris_table[1][1]
petal_lenght=iris_table[2][1]
petal_width =iris_table[3][1]
label       =iris_table[4][1]

training = []
for index in 0 ... sepal_length.size
  training.push [sepal_length[index],sepal_width[index],petal_lenght[index],petal_width[index],label[index]]
end

dec_tree = DecisionTree::ID3Tree.new(attributes, training , 'Iris-setosa', sepal_length: :continuous, sepal_width: :continuous,petal_lenght: :continuous, petal_width: :continuous)
```

```ruby
time = Benchmark.measure do
  (0..10).each {
    dec_tree.train
  }
end

puts (time.real/10)*1000
```

```ruby
time = Benchmark.measure do
  (0..10).each {
    predicted = dec_tree.predict([5,2,6,1])
  }
end

puts predicted
puts (time.real/10)*1000
```
